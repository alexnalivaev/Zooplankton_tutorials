```python
# OK ICI TU DÉFINIS SURTOUT LES LONMIN, LONMAX DE L'INDIEN SUD
min_lon=20
max_lon=120
max_lat=90
min_lat=30

dim=(max_lat-min_lat,max_lon-min_lon+1)
print(dim)
```

### Une manière de boucler sur les modèles pour calculer une moyenne:

- Glob pour lire des fichiers contenus dans un répertoire:

Glob permet de trouver, dans un répertoire donné, des fichiers qui contiennent une séquence (en l'occurrence le nom d'un modèle).
On utilise * comme un raccourci qui signifie "toute séquence de caractères".
Par exemple "monmean_1995-2014_SO_zooc_Oyr_"+model+"_historical_*_gr_185001-201412.nc"
signifie qu'on va lire tous les fichiers qui commencent par "monmean_1995-2014_SO_zooc_Oyr_", puis contiennent un nom de modèle, puis contiennent "_historical_", puis toute séquence de caractères suivie de "_gr_185001-201412.nc".

- np.nansum si on veut intégrer (sommer) les données sur la verticale

```python
model_avg=np.zeros(dim)

for i,model in enumerate(list_of_models):
    #print(model)
    historical_file= glob(rep_data+"monmean_1995-2014_SO_zooc_Oyr_"+model+"_historical_*_gr_185001-201412.nc")[0]
    #print(historical_file) # si besoin pour vérifier qu'on lit bien les bons fichiers
    historical_data=Dataset(historical_file,'r',format='NETCDF4')
    
    #print(historical_data.variables)
    
    # Lire les variables
    lon=historical_data.variables['lon'][:]
    lat=historical_data.variables['lat'][:]
    
    zooc=historical_data.variables['zooc'][:,:,:,:]
    
    #Option: sélectionner sur l'indien Sud plutôt que tout l'océan Austral
    ilon_min,ilon_max=np.searchsorted(lon[:],[min_lon,max_lon])
    
    # Une manière pour intégrer les données sur la verticale: np.nansum au lieu de nanmean
    zooc_vertical_sum=np.nansum(zooc[0,:,:,ilon_min:ilon_max+1],axis=0)
    print(np.shape(zooc_vertical_sum))
    
    model_avg+=zooc_vertical_sum
model_avg=model_avg/len(list_of_models)
```
