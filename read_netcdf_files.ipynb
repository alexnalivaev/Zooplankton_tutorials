# Lire les variables d'un fichier NETCDF

##### Written April 9 2026

Objectif: lire les variables contenues dans un fichier CMIP6.

Les sorties CMIP6 sont sauvegardées au format NETCDF: ici on utilise le module netCDF4 de Python pour lire ces fichiers.

### Importer les modules nécessaires

```python
import numpy as np # NUMPY = numerical Python: module qui gère beaucoup de calculs de base

import netCDF4 as nc
from netCDF4 import Dataset
# le module netCDF4 et sa fonction Dataset nous permettent de lire un fichier NETCDF et les variables qu'il contient
```

### Définir le répertoire de stockage du fichier

```python
# Ici il faudra que tu changes rep_file pour ton répertoire local de stockage,
# par exemple /usr/cfievre/Documents si je me souviens bien
rep_file='/home/analivaev/Documents/THESE/Data/CMIP6/zooplankton_stage_Celeste/Test_files/'
test_file='timmean_SO_zooc_Oyr_IPSL-CM6A-LR_ssp585_r1i1p1f1_gr_208101-210012.nc'
```

Ci-dessous on utilise la fonction Dataset pour lire le fichier test_file, contenu dans le répertoire rep_file.
On obtient un objet de type netCDF4.Dataset qui stocke plusieurs variables.

```python
test_data=Dataset(rep_file+test_file,'r',format='NETCDF4')
```

```python
print(type(test_data)) # pour afficher la classe/le type d'un objet
```

Pour afficher l'ensemble des variables contenues dans un objet netCDF4.Dataset:

```python
print(test_data.variables)
```

Le fichier contient:
- une variable 'time': ici le fichier est une moyenne sur 20 ans, donc la dimension de la coordonnée temps est 1, c'est-à-dire qu'on a une seule entrée selon l'axe temps (t).
- une variable 'lon' et une variable 'lat': les coordonnées longitude et latitude (voir ci-dessous) (axes x et y).
- une variable 'olevel' qui désigne la profondeur (axe z)

- une variable 'zooc' (celle qui nous intéresse, qui décrit le zooplancton), elle est de dimensions (t,z,y,x).

La variable zooc décrit la biomasse de zooplancton (concentration, en $mol.m^{-3}$, exprimée en "moles de Carbone") et est 4D: pour chaque pixel de longitude, latitude et profondeur, et chaque pas de temps (en l'occurrence un seul), on a une valeur. 

### On veut regarder plus précisément les variables

On va commencer par les longitudes et les latitudes: les variables sont des objets netCDF4.Variable, on peut les afficher comme ci-dessous.
On a 360 valeurs de longitude, exprimées en °E, et 60 valeurs de latitude (en °N - comme on est dans l'hémisphère sud les valeurs sont négatives).

```python
test_data.variables['lon']
```

```python
test_data.variables['lat']
```

Pour afficher les valeurs de longitude, latitude: on ajoute '[:]' après les variables.

Les longitudes sont grillées tous les 1°, et vont de -180° à 180°: dans ce fichier test, on a des données sur toutes les longitudes du globe.
Les latitudes sont de même grillées tous les 1° et vont de -89.5°N (pôle Sud) à -30.5°N (limite Nord de l'Océan Austral selon la définition que j'utilise).
Le fichier contient des données de zooplancton restreintes à l'Océan Austral (j'ai extrait ces coordonnées géographiques des sorties de modèle initiales).

```python
print(test_data.variables['lon'][:])
print(test_data.variables['lat'][:])
```

On veut stocker les variables lon, lat dans un objet qui sera plus facile à appeler: on définit une variable lon et une variable lat.

```python
lon=test_data.variables['lon'][:]
lat=test_data.variables['lat'][:]
```

On regarde la variable 'olevel' qui indique la profondeur.

```python
test_data.variables['olevel']
```

```python
z=test_data.variables['olevel'][:]
print(z[:])
```

### Et la variable 'zooplancton':

```python
test_data.variables['zooc']
```

```python
zooc=test_data.variables['zooc'][:,:,:,:]
print(zooc)
```

La variable zooc est une matrice 4D: pour extraire des données sur un pixel en particulier:

```python
print(zooc[0,0,20,10])
# ici on regarde la variable zooc pour, dans l'ordre:
# la première valeur selon la coordonnée temps (Python commence à compter à 0)
# la première valeur selon la coordonnée profondeur - Donc la surface
# la 20e valeur selon la coordonnée latitude (choisie au hasard)
# la 10e selon la coordonnée longitude (idem)
```

### Masked Arrays

```python
print(type(zooc))
```

Les variables qu'on a extraites sont des Arrays, ou des vecteurs et des matrices.
Plus précisément ce sont des MaskedArray en Python: ce sont des arrays dont les valeurs sont potentiellement masquées.

Par exemple, la variable 'zooc' pour le zooplancton est masquée au dessus des continents, sur les points Terre (notamment sur le continent Antarctique).

Voir la documentation:
https://numpy.org/doc/stable/reference/maskedarray.generic.html

```python
#Lorsque la variable est masquée:
print(zooc[0,0,0,0])   
```

Ici on lui demande d'afficher une valeur à la première latitude (-89.5°N, sur le continent Antarctique).
